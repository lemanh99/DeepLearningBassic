```
from tensorflow.keras.models import  Sequential
from tensorflow.keras.layers import  Dense, Dropout, Activation, Flatten, Dropout
from tensorflow.keras.layers import  Conv2D, MaxPool2D
```

```
input_shape = (224, 224, 3)
model = None
```

```
def create_model(num_layers, filters):
  global model
  for num_layer in range(num_layers):
    model.add(Conv2D(filters=filters, kernel_size=3,strides=1,padding='same',  activation='relu'))
  model.add(MaxPool2D(pool_size=(2, 2)))
```

```
filters=64
model = Sequential()
model.add(Conv2D(filters=64, kernel_size=3,strides=1, padding='same',  activation='relu', input_shape = input_shape))
for i in range(1, 6):
  if i<3:
    create_model(i, filters)
  else:
    create_model(3, filters)
  if i<4:
    filters*=2

model.add(Flatten())
model.add(Dense(4096, activation='relu'))
model.add(Dense(4096, activation='relu'))
model.add(Dense(1000, activation='softmax'))
model.summary()
```
